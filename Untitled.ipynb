```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import re
import csv
import datetime
from tqdm import tqdm
from matplotlib.patches import Polygon
%matplotlib inline
```

```python

#dates = pickle.load( open( "dates.p", "rb"))
#duration = pickle.load( open( "duration.p", "rb"))
#dbc = pickle.load( open("days_between_clicks.p", "rb"))
experience = pickle.load( open("experience.p", "rb"))
salary = pickle.load( open("salary.p", "rb"))
companyRating = pickle.load( open("companyRating.p", "rb"))
wordCount = pickle.load( open("wordCount.p", "rb"))
jobAgeCount = pickle.load( open("jobage.p", "rb"))
clicks = pickle.load( open("clicks.p", "rb"))
#hoc = pickle.load( open("hoc.p", "rb"))
#category = pickle.load( open("category.p", "rb"))
```

```python
count = 0
x1 = []
x2 = []
x3 = []
x4 = []
y1 = []
for d in tqdm(clicks.keys()):
    count += 1
    if jobAgeCount[d] != 0 and int(wordCount[d]) > 20 and salary != 0:
        x1.append(float(experience[d]))
        y1.append(clicks[d]/jobAgeCount[d])
        x4.append(float(companyRating[d]))
        x2.append(int(wordCount[d]))
        x3.append(int(salary[d]))  
    #int(re.search(r'\d+', str(d)).group())
```

```python
#fig, axes = plt.subplots(4, sharey=True)
#fig.set_size_inches(18.5, 20, forward=True)
plt.figure(figsize=(20,15))
plt.scatter(x1,y1, s=30, alpha=0.15, marker='o')
plt.plot(np.unique(x1), np.poly1d(np.polyfit(x1, y1, 1))(np.unique(x1)))
plt.title("Hotness vs Experience")
plt.show()
plt.close()
plt.figure(figsize=(20,15))
plt.scatter(x2,y1, s=30, alpha=0.15, marker='o')
plt.plot(np.unique(x2), np.poly1d(np.polyfit(x2, y1, 1))(np.unique(x2)))
plt.title("Hotness vs Word Count")
plt.show()
plt.close()
plt.figure(figsize=(20,15))
plt.scatter(x3,y1, s=30, alpha=0.15, marker='o')
plt.plot(np.unique(x3), np.poly1d(np.polyfit(x3, y1, 1))(np.unique(x3)))
plt.title("Hotness vs Salary")
plt.show()
plt.close()
plt.figure(figsize=(20,15))
plt.scatter(x4,y1, s=30, alpha=0.15, marker='o')
plt.plot(np.unique(x4), np.poly1d(np.polyfit(x4, y1, 1))(np.unique(x4)))
plt.title("Hotness vs Ratings")
plt.show()

#m1,b1 = np.polyfit(x1,y1,1)
#plt.plot(x1,m1*x1+b1,'-')
#axes[1].scatter(x2,y1, s=30, alpha=0.15, marker='o')
#axes[2].scatter(x3,y1, s=30, alpha=0.15, marker='o')
#axes[3].scatter(x4,y1, s=30, alpha=0.15, marker='o')
```
